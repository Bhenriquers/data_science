```python
import pandas as pd
```

```python
dados = pd.read_json('/content/dados_vendas_clientes.json')
```

```python
dados.head()
```

```python
dados = pd.json_normalize(dados['dados_vendas'])
```

```python
colunas = list(dados.columns)
```

```python
colunas
```

```python
dados = dados.explode(colunas[1:])
```

```python
dados
```

```python
dados.info()
```

```python
import numpy as np
```

```python
dados['Valor da compra'] = dados['Valor da compra'].apply(lambda x: x.replace('R$','').replace(',', '.').strip())
```

```python
dados
```

```python
dados['Valor da compra'] = dados['Valor da compra'].astype(np.float64)
```

```python
dados['Cliente'] = dados['Cliente'].str.lower()
```

```python
dados['Cliente'] = dados['Cliente'].str.replace(r"[@_]", " ", regex=True).str.strip()
```

```python
dados[['Nome Cliente', 'Cliente ID']] = dados['Cliente'].str.rsplit(' ', n=1, expand=True)
```

```python
dados.reset_index(drop = True, inplace = True)
```

```python
dados.drop(columns=['Cliente'], inplace = True)
```

```python
dados.head(5)
```

```python
dados['Data de venda'] = pd.to_datetime(dados['Data de venda'])
```
